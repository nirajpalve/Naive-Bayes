```python
import pandas as pd       # Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
Train_Data = pd.read_csv("SalaryData_Train.CSV")     # Loading Train - DataSet
Train_Data
```

```python
Train_Data.shape    # Dimension of test data
```

```python
Train_Data.dtypes
```

```python
Train_Data.describe(include = "all")
```

```python
Train_Data.describe(include = "all").nunique()
```

```python
Label = LabelEncoder()   
```

```python
Train_Data['workclass'] = Label.fit_transform(Train_Data['workclass'])
Train_Data['education'] = Label.fit_transform(Train_Data['education'])
Train_Data['maritalstatus'] = Label.fit_transform(Train_Data['maritalstatus'])
Train_Data['occupation'] = Label.fit_transform(Train_Data['occupation'])
Train_Data['relationship'] = Label.fit_transform(Train_Data['relationship'])
Train_Data['race'] = Label.fit_transform(Train_Data['race'])
Train_Data['sex'] = Label.fit_transform(Train_Data['sex'])
Train_Data['native'] = Label.fit_transform(Train_Data['native'])
Train_Data['Salary'] = Label.fit_transform(Train_Data['Salary'])
```

```python
Train_Data
```

```python
Train_Data.dtypes
```

# Model Building

```python
X = Train_Data.drop(['Salary'], axis = 1)
Y = Train_Data['Salary']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.20, random_state = 42)
```

```python
X_Train.shape, Y_Train.shape
```

```python
X_Test.shape, Y_Test.shape
```

## Model Training & Testing

```python
from sklearn.naive_bayes import MultinomialNB as MB
```

```python
classifier_mb = MB()
```

```python
classifier_mb.fit(X_Train, Y_Train)
MB()
```

```python
Y_pred = classifier_mb.predict(X_Test)
Y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, classification_report
```

```python
print("Accurracy Score:", round(accuracy_score(Y_Test, Y_pred), 4))
```

```python
print("Precision Score:", round(precision_score(Y_Test, Y_pred), 4))
```

```python
print("Confusion Matrix:", confusion_matrix(Y_Test, Y_pred))
```

```python
print("Classification Report: ", classification_report(Y_Test, Y_pred))
```
